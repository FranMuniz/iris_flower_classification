### Graph Plotting

```python
# Relationship between sepal length and width

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)

# Placing legend outside the figure

plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.show()
```

```python
# Relationship between petal length and width

sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)

# Placing legend outside the figure

plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.show()
```

```python
# Graphically represent all the relationships of the df

sns.pairplot(df.drop(['Id'], axis=1), hue='Species', height=2)
```

```python
# Distribution of data in columns

fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title('Sepal Length')
axes[0,0].hist(df['SepalLengthCm'], bins=7)

axes[0,1].set_title('Sepal Width')
axes[0,1].hist(df['SepalWidthCm'], bins=5)

axes[1,0].set_title('Petal Length')
axes[1,0].hist(df['PetalLengthCm'], bins=6)

axes[1,1].set_title('Petal Width')
axes[1,1].hist(df['PetalWidthCm'], bins=6)
```

```python
# Plotting the distributions for 'SepalLengthCm'
plot = sns.FacetGrid(df, hue='Species', col='Species', col_wrap=2, height=4, aspect=1.2) 
plot.map(sns.histplot, 'SepalLengthCm', kde=True) 
plot.add_legend()
plt.suptitle('Distribution of SepalLengthCm by Species', y=1.02) 

# Plotting the distributions for 'SepalWidthCm'
plot = sns.FacetGrid(df, hue='Species', col='Species', col_wrap=2, height=4, aspect=1.2)
plot.map(sns.histplot, 'SepalWidthCm', kde=True)
plot.add_legend()
plt.suptitle('Distribution of SepalWidthCm by Species', y=1.02)

# Plotting the distributions for 'PetalLengthCm'
plot = sns.FacetGrid(df, hue='Species', col='Species', col_wrap=2, height=4, aspect=1.2)
plot.map(sns.histplot, 'PetalLengthCm', kde=True)
plot.add_legend()
plt.suptitle('Distribution of PetalLengthCm by Species', y=1.02)

# Plotting the distributions for 'PetalWidthCm'
plot = sns.FacetGrid(df, hue='Species', col='Species', col_wrap=2, height=4, aspect=1.2)
plot.map(sns.histplot, 'PetalWidthCm', kde=True)
plot.add_legend()
plt.suptitle('Distribution of PetalWidthCm by Species', y=1.02)

plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust the layout to prevent title overlap
plt.show()
```

```python
# Correlation analysis of column pairs of df

correlation_matrix = data.corr(method='pearson', numeric_only=True)
print(correlation_matrix)
```

```python
# Heat map

plt.figure(figsize=(8, 6)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Numeric Attribute Correlation Heatmap')
plt.show()
```

```python
def graph(y):
    sns.boxenplot(x='Species', y=y, data=df)

plt.figure(figsize=(10,10))

# Adding the subplot at the specified grid position

plt.subplot(221)
graph('SepalLengthCm')

plt.subplot(222)
graph('SepalWidthCm')

plt.subplot(223)
graph('PetalLengthCm')

plt.subplot(224)
graph('PetalWidthCm')
```
